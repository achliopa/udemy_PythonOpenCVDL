```python
rainbow = cv2.imread('../DATA/rainbow.jpg')
show_rainbow = cv2.cvtColor(rainbow,cv2.COLOR_BGR2RGB)
```

```python
img = rainbow
```

```python
img.shape
```

```python
mask = np.zeros(img.shape[:2],np.uint8)
```

```python
# plt.imshow(mask,cmap='gray')
mask[300:400,100:400] = 255
```

```python
plt.imshow(mask,cmap='gray')
```

```python
masked_img = cv2.bitwise_and(img,img,mask=mask)
plt.imshow(masked_img)
```

```python
show_masked_img = cv2.bitwise_and(show_rainbow,show_rainbow,mask=mask)
plt.imshow(show_masked_img)
```

```python
hist_mask_values_red = cv2.calcHist([rainbow],channels=[2],mask=mask,histSize=[256],ranges=[0,256])
```

```python
hist_values_red = cv2.calcHist([rainbow],channels=[2],mask=None,histSize=[256],ranges=[0,256])
```

```python
plt.plot(hist_mask_values_red)
plt.title('RED HISTOGRAM FOR MASKED IMAGE')
```

```python
plt.plot(hist_values_red)
plt.title('RED HISTOGRAM FOR COMPLETE IMAGE')
```

```python
gorilla = cv2.imread('../DATA/gorilla.jpg',0)
```

```python
def display_img(img,cmap=None):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap)
```

```python
display_img(gorilla,cmap='gray')
```

```python
hist_values = cv2.calcHist([gorilla],channels=[0],mask=None,histSize=[256],ranges=[0,256])
```

```python
plt.plot(hist_values)
```

```python
eq_goriila = cv2.equalizeHist(gorilla)
```

```python
display_img(eq_goriila,cmap='gray')
```

```python
eq_hist_values = cv2.calcHist([eq_goriila],channels=[0],mask=None,histSize=[256],ranges=[0,256])
```

```python
plt.plot(eq_hist_values)
```

```python
color_gorilla = cv2.imread('../DATA/gorilla.jpg')
show_gorilla = cv2.cvtColor(color_gorilla,cv2.COLOR_BGR2RGB)
display_img(show_gorilla)
```

```python
hsv_gorilla = cv2.cvtColor(color_gorilla,cv2.COLOR_BGR2HSV)
value_channel = hsv_gorilla[:,:,2]
eq_value_channel = cv2.equalizeHist(value_channel)
hsv_gorilla[:,:,2] = eq_value_channel
show_eq_color_gorilla = cv2.cvtColor(hsv_gorilla,cv2.COLOR_HSV2RGB)
display_img(show_eq_color_gorilla)
```
